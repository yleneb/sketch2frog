```python
import pathlib
import zipfile
import IPython
import random

PATH = pathlib.Path.cwd()
DATA_PATH = PATH.parent / "data"
ZIP_PATH = DATA_PATH / 'data-224.zip'
```

https://github.com/jonshamir/frog-dataset contains >7k images of frogs on a white background. Using DownGit we can download just the folder of processed 224x224 images.

https://minhaskamal.github.io/DownGit/#/home?url=https://github.com/jonshamir/frog-dataset/tree/master/data-224

```python
# unzip folder into data/raw
with zipfile.ZipFile(ZIP_PATH, 'r') as zip_ref:
    zip_ref.extractall(DATA_PATH / 'raw')

# delete the zip file
# os.remove(ZIP_PATH)
```

```python
filepaths = [x for x in DATA_PATH.glob('raw/*/*.png')]

IPython.display.Image(filename=random.choice(filepaths))
```
